```python
from docx import Document
document = Document('./data/law.docx')
print(f'document == {dir(document)}')
full_text = ''
for index, para in enumerate(document.paragraphs) :
    print(f'paragraph: {para.text}')
    full_text += f'{para.text}\n'
full_text
```

```python
# full_text 문서쪼개기
%pip install tiktoken

import tiktoken


def split_text(full_text, chunk_size) :
    encoder = tiktoken.encoding_for_model('gpt-4o')
    total_encoding = encoder.encode(full_text)
    total_token_count = len(total_encoding)
    text_list = []
    for i in range(0, total_token_count, chunk_size) :
        chunk = total_encoding[i:i+chunk_size]
        decoded = encoder.decode(chunk)
        text_list.append(decoded)
    return text_list

chunk_list = split_text(full_text, 500)
chunk_list
```

```python
%pip install chromadb

import chromadb
chroma_client = chromadb.Client()
collection_name = 'law_collection'
law_collection = chroma_client.create_collection(collection_name)

```

```python
import os
from dotenv import load_dotenv
from chromadb.utils.embedding_functions import OpenAIEmbeddingFunction


load_dotenv()
open_api_key = os.getenv('OPENAI_API_KEY')
openai_embedding = OpenAIEmbeddingFunction(api_key = open_api_key, model_name = 'text-embedding-3-large')

```

```python
# 강의자료는 get_or_create_collection 옵션으로 embedding_function=openai_embedding가 있는데, 실제로는 conflict이 나타나 없애주면 된다.
law_collection = chroma_client.get_or_create_collection(collection_name ) 

# law_collection에 삽입할때, ids가 인의적으로 생성해주고 저장을 같이해줘야함.
id_list = []
for idx, chunk in enumerate(chunk_list) :
    # print(f'idx: {idx} / chunk: {chunk}')
    id_list.append(f'{idx}' * 3)

# C:\Users\USER\.cache\chroma\onnx_models\all-MiniLM-L6-v2\onnx.tar.gz: 100%|██████████| 79.3M/79.3M [00:21<00:00, 3.82MiB/s]  
law_collection.add(documents = chunk_list, ids = id_list)
```

```python
query = '원자력안전법 1조에 대한 내용을 요약해서 알려주세요'

retrieved_doc_list = law_collection.query(query_texts = query, n_results=3)

retrieved_doc_list['documents'][0]
```
